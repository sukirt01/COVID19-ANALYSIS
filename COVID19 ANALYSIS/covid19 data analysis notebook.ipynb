### Import the modules 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

### importing covid19 dataset
importing "Covid19_Confirmed_dataset.csv" from "./Dataset" folder. 


```python
data_corona = pd.read_csv("covid19_Confirmed_dataset.csv")
data_corona.head(10)
```

```python
data_corona.shape
```

### Deleting the useless columns

```python
data_corona.drop(["Lat","Long"],axis=1)
```

```python
data_corona.head(10)
```

### Aggregating the rows by the country

```python
data = data_corona.groupby("Country/Region").sum()
```

```python
data.head()
```

```python
data.shape
```

### Visualizing data related to a country 


```python
data.loc["India"].plot()
data.loc["Italy"].plot()
data.loc["Spain"].plot()
plt.legend()
```

### Task3: Calculating a good measure 
 describing the spread of the virus in a country. 

```python
data.loc['China'][:3].plot()
```

```python
data.loc['Italy'].plot()
```

### task 3.1: caculating the first derivative of the curve

```python
data.loc["China"].diff().plot()
```

### finding maxmimum infection rate for China

```python
data.loc["China"].diff().max()
```

```python
data.loc["Italy"].diff().max()
```

```python
data.loc["Spain"].diff().max()
```

### finding maximum infection rate for all of the countries. 

```python
countries = list(data.index)
max_inf_rate = []

for c in countries:
    max_inf_rate.append(data.loc[c].diff().max())   
```

```python
data["max_inf_rate"] = max_inf_rate
data.head()
```

### creating a new dataframe with only needed column 

```python
corona_analysis = pd.DataFrame(data["max_inf_rate"])
```

```python
corona_analysis.head()
```
